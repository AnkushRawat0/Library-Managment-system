```python
import pandas as pd
import numpy as np 
import random 
from datetime import datetime , timedelta
```

```python
n= 1000
```

```python
book_titles = ['DSA', 'Python', 'AI Basics', 'Machine Learning', 'DBMS', 'Operating Systems']
categories = ['Tech', 'Programming', 'AI', 'Data Science']
users = range(1,201) 

start_date = datetime(2025 ,1,1)
```

```python
data = [] 
```

```python
for i in range(n):
    issue_date = start_date + timedelta(days = random.randint(0,90))

    #return after 1-20 days
    return_days = random.randint(1,20)
    return_date = issue_date + timedelta(days = return_days)

    #latefine logic 
    fine  = 0 
    if return_days  > 10:
        fine = (return_days  -10 ) * 10

    data.append([
        i+1 ,
        random.choice(book_titles),
        random.choice(categories),
        random.choice(users),
        issue_date,
        return_date,
        fine
        
    ])

df = pd.DataFrame(data , columns=[
    'transaction_id', 'book_title', 'category', 'user_id',
    'issue_date', 'return_date', 'fine_amount'
])     
```

```python
df.to_csv("../data/library_data.csv" , index =False)

df.head()
```
